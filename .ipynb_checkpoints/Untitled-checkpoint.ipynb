```python

# Import necessary libraries
import pandas as pd
import numpy as np
from prophet import Prophet
import requests
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import seaborn as sns
import yfinance as yf

# Function to fetch stock data from Yahoo Finance using yfinance
def fetch_stock_data(symbol, start_date='2020-01-01'):
    df = yf.download(symbol, start=start_date)
    df = df[['Adj Close']]
    df.reset_index(inplace=True)
    df.columns = ['ds', 'y']
    return df

# Fetch stock data
symbol = 'AAPL'  # Example stock symbol (Apple Inc.)
stock_data = fetch_stock_data(symbol)

# Exploratory Data Analysis (EDA)
# Plotting the time series data
plt.figure(figsize=(10, 6))
plt.plot(stock_data['ds'], stock_data['y'])
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title(f'{symbol} Stock Price Over Time')
plt.show()

# Checking for missing values
print(stock_data.isna().sum())

# Checking basic statistics
print(stock_data.describe())

# Preprocessing
# Handling missing values (if any)
stock_data = stock_data.dropna()

# Ensuring the data is sorted by date
stock_data = stock_data.sort_values(by='ds')

# Feature engineering (if necessary)
# For Prophet, we typically only need 'ds' and 'y', so no additional features are added

# Split data into training and test sets
train_size = int(len(stock_data) * 0.8)
train_data = stock_data[:train_size]
test_data = stock_data[train_size:]

# Initialize and fit the Prophet model
model = Prophet()
model.fit(train_data)

# Make future dataframe and predict
future = model.make_future_dataframe(periods=len(test_data))
forecast = model.predict(future)

# Plot forecast
fig1 = model.plot(forecast)
plt.title(f'{symbol} Stock Price Forecast')
plt.show()

# Plot actual vs predicted
plt.figure(figsize=(10, 6))
plt.plot(test_data['ds'], test_data['y'], label='Actual')
plt.plot(test_data['ds'], forecast['yhat'].iloc[-len(test_data):].values, label='Predicted')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{symbol} Stock Price: Actual vs Predicted')
plt.legend()
plt.show()

# Evaluate forecast
y_true = test_data['y'].values
y_pred = forecast['yhat'].values[-len(test_data):]

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = mean_absolute_percentage_error(y_true, y_pred)
squared_error = mean_squared_error(y_true, y_pred, squared=True)

print(f'RMSE: {rmse}')
print(f'MAPE: {mape}')
print(f'Squared Error: {squared_error}')
```

```python

# Import necessary libraries
import pandas as pd
import numpy as np
import yfinance as yf
from prophet import Prophet
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import pmdarima as pm
from xgboost import XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler

# Fetch stock data from Yahoo Finance using yfinance
def fetch_stock_data(symbol, start_date='2020-01-01'):
    df = yf.download(symbol, start=start_date)
    df = df[['Adj Close']]
    df.reset_index(inplace=True)
    df.columns = ['ds', 'y']
    return df

# Fetch stock data
symbol = 'AAPL'  # Example stock symbol (Apple Inc.)
stock_data = fetch_stock_data(symbol)

# Exploratory Data Analysis (EDA)
# Plotting the time series data
plt.figure(figsize=(10, 6))
plt.plot(stock_data['ds'], stock_data['y'])
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title(f'{symbol} Stock Price Over Time')
plt.show()

# Checking for missing values
print(stock_data.isna().sum())

# Checking basic statistics
print(stock_data.describe())

# Preprocessing
# Handling missing values (if any)
stock_data = stock_data.dropna()

# Ensuring the data is sorted by date
stock_data = stock_data.sort_values(by='ds')

# Split data into training and test sets
train_size = int(len(stock_data) * 0.8)
train_data = stock_data[:train_size]
test_data = stock_data[train_size:]

# Helper function to evaluate the models
def evaluate_model(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mean_absolute_percentage_error(y_true, y_pred)
    squared_error = mean_squared_error(y_true, y_pred, squared=True)
    print(f'{model_name} - RMSE: {rmse}, MAPE: {mape}, Squared Error: {squared_error}')
    return rmse, mape, squared_error

# Prophet Model
model_prophet = Prophet()
model_prophet.fit(train_data)
future = model_prophet.make_future_dataframe(periods=len(test_data))
forecast_prophet = model_prophet.predict(future)

y_true = test_data['y'].values
y_pred_prophet = forecast_prophet['yhat'].values[-len(test_data):]
rmse_prophet, mape_prophet, se_prophet = evaluate_model(y_true, y_pred_prophet, 'Prophet')

# ARIMA Model
model_arima = pm.auto_arima(train_data['y'], seasonal=False)
forecast_arima = model_arima.predict(n_periods=len(test_data))

y_pred_arima = forecast_arima
rmse_arima, mape_arima, se_arima = evaluate_model(y_true, y_pred_arima, 'ARIMA')

# LSTM Model
def create_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Prepare data for LSTM
def prepare_data_lstm(data, time_steps=1):
    X, y = [], []
    for i in range(len(data) - time_steps - 1):
        X.append(data[i:(i + time_steps), 0])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(stock_data[['y']].values)
train_scaled = scaled_data[:train_size]
test_scaled = scaled_data[train_size:]

time_steps = 60
X_train, y_train = prepare_data_lstm(train_scaled, time_steps)
X_test, y_test = prepare_data_lstm(test_scaled, time_steps)

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

model_lstm = create_lstm_model((X_train.shape[1], 1))
model_lstm.fit(X_train, y_train, epochs=20, batch_size=32)

predicted_lstm = model_lstm.predict(X_test)
predicted_lstm = scaler.inverse_transform(predicted_lstm)

y_test = scaler.inverse_transform([y_test])
y_pred_lstm = predicted_lstm[:, 0]
rmse_lstm, mape_lstm, se_lstm = evaluate_model(y_test[0], y_pred_lstm, 'LSTM')

# XGBoost Model
train_data_xgb = train_data.copy()
train_data_xgb['y_shifted'] = train_data_xgb['y'].shift(-1)
train_data_xgb.dropna(inplace=True)

X_train_xgb = train_data_xgb.drop(['ds', 'y_shifted'], axis=1)
y_train_xgb = train_data_xgb['y_shifted']
X_test_xgb = test_data.drop(['ds', 'y'], axis=1)

model_xgb = XGBRegressor(objective='reg:squarederror')
model_xgb.fit(X_train_xgb, y_train_xgb)
y_pred_xgb = model_xgb.predict(X_test_xgb)

rmse_xgb, mape_xgb, se_xgb = evaluate_model(y_true, y_pred_xgb, 'XGBoost')

# ARIMA-LSTM Hybrid Model
train_arima_residuals = y_train - model_arima.predict_in_sample()
model_lstm_residual = create_lstm_model((time_steps, 1))
model_lstm_residual.fit(X_train, train_arima_residuals, epochs=20, batch_size=32)

predicted_arima_lstm = model_lstm_residual.predict(X_test)
predicted_arima_lstm = scaler.inverse_transform(predicted_arima_lstm)
residuals_forecast = model_arima.predict(n_periods=len(test_data)) + predicted_arima_lstm[:, 0]

rmse_arima_lstm, mape_arima_lstm, se_arima_lstm = evaluate_model(y_true, residuals_forecast, 'ARIMA-LSTM Hybrid')

# Prophet-LSTM Hybrid Model
forecast_prophet_residuals = train_data['y'] - forecast_prophet['yhat'][:len(train_data)]
model_lstm_residual.fit(X_train, forecast_prophet_residuals.values, epochs=20, batch_size=32)

predicted_prophet_lstm = model_lstm_residual.predict(X_test)
predicted_prophet_lstm = scaler.inverse_transform(predicted_prophet_lstm)
residuals_forecast_prophet = forecast_prophet['yhat'].values[-len(test_data):] + predicted_prophet_lstm[:, 0]

rmse_prophet_lstm, mape_prophet_lstm, se_prophet_lstm = evaluate_model(y_true, residuals_forecast_prophet, 'Prophet-LSTM Hybrid')

# Plot actual vs predicted for the best model
best_model_name = min(
    ['Prophet', 'ARIMA', 'LSTM', 'XGBoost', 'ARIMA-LSTM Hybrid', 'Prophet-LSTM Hybrid'], 
    key=lambda x: locals()[f'rmse_{x.lower().replace("-", "_")}']
)
y_pred_best = locals()[f'y_pred_{best_model_name.lower().replace("-", "_")}']

plt.figure(figsize=(10, 6))
plt.plot(test_data['ds'], y_true, label='Actual')
plt.plot(test_data['ds'], y_pred_best, label=f'Predicted ({best_model_name})')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{symbol} Stock Price: Actual vs Predicted ({best_model_name})')
plt.legend()
plt.show()

best_model_rmse = locals()[f'rmse_{best_model_name.lower().replace("-", "_")}']
print(f'The best model is {best_model_name} with RMSE: {best_model_rmse}')
```
